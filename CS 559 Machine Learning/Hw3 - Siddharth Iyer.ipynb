# Problem 4: Part II

## (1) - Classification of Points (Train and Test)

```python
import pandas as pd
import numpy as np
from sklearn.neighbors import KNeighborsClassifier

train = pd.read_csv("train.csv")
test = pd.read_csv("test.csv")

X_train = train[["x", "y", "z"]]
y_train = train[["class"]]

X_test = test[["x", "y", "z"]]
y_test = test[["actual-class"]]

knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train, np.ravel(y_train))

prediction = knn.predict(X_test)
probs = knn.predict_proba(X_test)

print("The predicted values for the test data: ")
out = test[["ID"]]
out["Prediction"] = prediction
display(out)

display(pd.DataFrame(probs))
```

## (2) - Euclidian Weighted Distance

```python
knn_weighted = KNeighborsClassifier(n_neighbors=3, weights="distance")
knn_weighted.fit(X_train, y_train)

weighted_prediction = knn_weighted.predict(X_test)
weighted_probs = knn_weighted.predict_proba(X_test)

print("The predicted values for the test data: ")
weighted_out = test[["ID"]]
weighted_out["Prediction"] = weighted_prediction
display(weighted_out)

display(pd.DataFrame(weighted_probs))
```

## (3) - Prediction Accuracy for both models

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score
from sklearn.metrics import f1_score

print("Accuracy: " + str(accuracy_score(y_test, prediction)))
print("F-measure: " + str(f1_score(y_test, prediction)))
print("Precision: " + str(precision_score(y_test, prediction)))
print("Confusion Matrix for unweighted prediction: ")
display(pd.DataFrame(confusion_matrix(y_test, prediction)))

print("Accuracy: " + str(accuracy_score(y_test, weighted_prediction)))
print("F-measure: " + str(f1_score(y_test, weighted_prediction)))
print("Precision: " + str(precision_score(y_test, weighted_prediction)))
print("Confusion matrix for weighted prediction: ")
display(pd.DataFrame(confusion_matrix(y_test, weighted_prediction)))
```
