# Import libraries
### Pandas, Sklearn, Numpy

```python
import pandas as pd
import numpy as np
from numpy import absolute

from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold

from sklearn.linear_model import ElasticNet, Lasso, Ridge
```

# Load dataset and split into input/output

```python
df = pd.read_excel("ENB2012_data.xlsx")
Y = df["Y2"]
X = df.drop("Y2", axis=1)
display(df)
```

# Perform 5-fold Cross Validation and score using MSE

```python
cv = KFold(n_splits = 5, shuffle=True)

enet = ElasticNet()
lasso = Lasso()
ridge = Ridge()

enet_scores = cross_val_score(enet, X, Y, scoring = "neg_mean_squared_error",cv=cv)
lasso_scores = cross_val_score(lasso, X, Y, scoring = "neg_mean_squared_error", cv=cv)
ridge_scores = cross_val_score(ridge, X, Y, scoring = "neg_mean_squared_error", cv=cv)

print("Elastic Net CV=5 Score: ", np.average(absolute(enet_scores)))
print("Lasso CV=5 Score: ", np.average(absolute(lasso_scores)))
print("Ridge CV=5 Score: ", np.average(absolute(ridge_scores)))
```

Based on the results above, Ridge regression has the lowest error; Elastic Net and Lasso are 2nd and 3rd best respectively.
